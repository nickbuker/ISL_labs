##  2.3.2 Graphics

```r
# plot() used to make scatter plots
x=rnorm(100)
y=rnorm(100)
plot(x, y, xlab='this is the x-axis', 
     ylab='this is the y-axix', main='Plot of x vs y')
```

```r
# Save image as png, pdf etc, dev.off() indicates done plotting
png('../images/chapter2.1.png')
plot(x, y, xlab='this is the x-axis', col='green',
     ylab='this is the y-axix', main='Plot of x vs y')
dev.off()
```

```r
# seq() creates vector and n1:n2 is shorthand
x = seq(1, 10)
x = 1:10
y = seq(-pi, pi, length=50)
```

```r
# contour() and image() used to create 3d plots (the latter is colored)
x = y
f = outer(x, y, function(x, y)cos(y)/(1+x^2))
contour(x, y, f)
# add layers the plots
contour(x, y, f, nlevels=45,add=TRUE)
fa = (f-t(f))/2
contour(x, y, fa, nlevels=15)
```

```r
image(x, y, fa)
persp(x, y, fa)
```

```r
persp(x, y, fa, theta=30)
png('../images/chapter2.2.png')
persp(x, y, fa, theta=30, phi=20)
dev.off()
```

```r
persp(x, y, fa, theta=30, phi=70, col='blue')
persp(x, y, fa, theta=30, phi=40)
```

## 2.3.3 Indexing Data

```r
# Create 4x4 matrix with range 1:16
A = matrix(1:16,4,4)
A
```

```r
# slice element from row 2 column 3, note that indexing starts from 1
A[2,3]
```

```r
# slice elements from rows 1 and 3 and columns 2 and 4
A[c(1,3),c(2,4)]
```

```r
# slice elements from rows 1-3 and colums 2-4 (inclusive)
A[1:3,2:4]
```

```r
# slice rows from 1-2 and all columns
A[1:2,]
```

```r
# slice all rows and columns from 1-2
A[,1:2]
```

```r
# slice row 1
A[1,]
```

```r
# grab rows other than 1 and 3 and all columns
A[-c(1,3),]
```

```r
# get shape of matrix
dim(A)
```

## 2.3.5 Additional Graphical and Numerical Summaries

```r
# plot Auto data
plot(Auto$cylinders, Auto$mpg)
```

```r
# use attach to make column names available in namespace
attach(Auto)
plot(cylinders, mpg)
```

```r
# cast cylinders to qualitative 
cylinders = as.factor(cylinders)
```

```r
# for categorical variables, box plots will be used automatically
plot(cylinders, mpg)
plot(cylinders, mpg, col='red')
plot(cylinders, mpg, col='red', varwidth=T)
plot(cylinders, mpg, col='red', varwidth=T, horizontal=T)
plot(cylinders, mpg, col='red', varwidth=T, xlab='cylinders', ylab='MPG')
```

```r
# save boxplots
png('../images/chapter2.3.bloxplots.png')
plot(cylinders, mpg, col='red', varwidth=T, xlab='cylinders', ylab='MPG')
dev.off()
```

```r
# make histograms
hist(mpg)
hist(mpg, col='red')
hist(mpg, col='red', breaks=15)
```

```r
# pairs makes a scatterplot matrix
pairs(Auto)
pairs(~mpg + displacement + horsepower + weight + acceleration, Auto)
```

```r
# indentify(x, y, label) used to show label when ploint clicked, doesn't show up in Jupyter
plot(horsepower, mpg)
identify(horsepower, mpg, name)
```

```r
# summary() create numerical summary of each variable
summary(Auto)
```
