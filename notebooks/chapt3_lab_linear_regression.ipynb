## Simple linear regression

```r
attach(Boston)
head(Boston)
# show column names
names(Boston)
```

```r
# fit model
lm.fit = lm(medv~lstat)
```

```r
# display information
lm.fit
```

```r
# show summary of stats
summary(lm.fit)
```

```r
# show pieces of information stored in lm.fit
names(lm.fit)
```

```r
# show confidence intervals
confint(lm.fit)
```

```r
# show confidence intervals
predict(lm.fit, data.frame(lstat = (c(5,10,15))), interval = 'confidence')
```

```r
# show prediction intervals
predict(lm.fit, data.frame(lstat = (c(5,10,15))), interval = 'prediction')
```

```r
plot(lstat, medv)
abline(lm.fit)
```

```r
# make some plots, change colors and markers
plot(lstat,medv,col='red')
abline(lm.fit, lwd=3, col='red')
plot(lstat,medv, pch=20)
abline(lm.fit, lwd=3)
plot(lstat,medv, pch="+")
abline(lm.fit, lwd=5)
plot(1:20,1:20,pch=(1:20))
```

```r
# par makes subplots
par(mfrow=c(2,2))
plot(lm.fit)
```

```r
plot(predict(lm.fit),residuals(lm.fit))
plot(predict(lm.fit),rstudent(lm.fit))
```

```r
# hat values used to comput leverage statistics
plot(hatvalues(lm.fit))
# which.max is like argmax
which.max(hatvalues(lm.fit))
```

## Multiple Linear Regression

```r
lm.fit = lm(medv~lstat + age, data=Boston)
summary(lm.fit)
```

```r
# use . to indicate use all features
lm.fit = lm(medv~., data=Boston)
summary(lm.fit)
```

```r
# find variance inflation factor
library(car)
vif(lm.fit)
```

```r
# use all features but age
lm.fit = lm(medv~.-age,data=Boston)
summary(lm.fit)
```

```r
# update model
lm.fit=update(lm.fit, ~.-age)
```

## Interaction terms

```r
# includes lstat, age and lstat * age
summary(lm(medv~lstat*age,data=Boston))
```

## Non-linear transformers of the predictors

```r
lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
```

```r
# ANOVA to determine degree to which quadratic model is improved
lm.fit=lm(medv~lstat)
anova(lm.fit, lm.fit2)
```

```r
par(mfrow=c(2,2))
plot(lm.fit)
par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
# fit polynomial model
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)
```

```r
# fit log model
summary(lm(medv~log(rm), data=Boston))
```
