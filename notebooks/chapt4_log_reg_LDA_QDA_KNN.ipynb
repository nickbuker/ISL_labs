## 4.6.2 Logistic regression

```r
# Fit model and assign family binomial
glm.fit=glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume, 
            data=Smarket,
            family=binomial)
summary(glm.fit)
```

```r
# Access the coefficients
coef(glm.fit)
```

```r
summary(glm.fit)$coef
```

```r
summary(glm.fit)$coef[,4]
```

```r
# Predict proba
contrasts(Direction)
glm.probs=predict(glm.fit, type='response')
glm.probs[1:10]
```

```r
# Create a vector of 1250 'Down' elements
glm.pred=rep('Down',1250)
# Convert the appropriate elements to up
glm.pred[glm.probs>0.5]='Up'
```

```r
# Generate confusion matrix
table(glm.pred, Direction)
```

```r
# Find the accuracy
(507+145)/1250
mean(glm.pred==Direction)
```

```r
# Create vector with 1250 boolean elements
train=(Year<2005)
# Use boolean masking to construct test data (Smarket.2005)
Smarket.2005=Smarket[!train,]
dim(Smarket.2005)
```

```r
Direction.2005=Direction[!train]
```

```r
# Fit model to training data
glm.fit=glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume,
            data=Smarket,
            family=binomial,
            subset=train)
# Make predictions on test
glm.probs=predict(glm.fit,
                  Smarket.2005,
                  type="response")
```

```r
# Construct vector of downs and then assign up where prob over 0.5
glm.pred=rep("Down",252)
glm.pred[glm.probs>0.5]="Up"
```

```r
# Create confusion matrix
table(glm.pred,Direction.2005)
```

```r
# Find the accuracy
mean(glm.pred==Direction.2005)
# Find the proportion inaccurately assigned
mean(glm.pred!=Direction.2005)
```

```r
# Remove features in order to attempt to improve performance
glm.fit=glm(Direction~Lag1+Lag2,
            data=Smarket,
            family=binomial,
            subset=train)
glm.probs=predict(glm.fit,
                 Smarket.2005,
                 type='response')
```

```r
glm.pred=rep('Down',252)
glm.pred[glm.probs>0.5]='Up'
table(glm.pred,Direction.2005)
```

```r
# Same accuracy as predicting an increase every day
mean(glm.pred==Direction.2005)
```

```r
# What if we only buy on days predicted to go up?
106/(106+76)
```

```r
predict(glm.fit,
        newdata=data.frame(Lag1=c(1.2,1.5),Lag2=c(1.1,-0.8)),
        type='response')
```

## 4.6.3 Linear discriminant analysis

```r
library(MASS)
```

```r
lda.fit=lda(Direction~Lag1+Lag2,
            data=Smarket,
            subset=train)
lda.fit
```

```r
plot(lda.fit)
```

```r
lda.pred=predict(lda.fit, Smarket.2005)
names(lda.pred)
```

```r
# Make confusion matrix
lda.class=lda.pred$class
table(lda.class,Direction.2005)
# Find accuracy
mean(lda.class==Direction.2005)
```

```r
# Recreate the predictions using the probabilities
sum(lda.pred$posterior[,1]>=0.5)
sum(lda.pred$posterior[,1]<0.5)
```

```r
# Probability output is prob of decrease
lda.pred$posterior[1:20,1]
lda.class[1:20]
```

```r
# Output of different threshold
sum(lda.pred$posterior[,1]>0.51)
```
